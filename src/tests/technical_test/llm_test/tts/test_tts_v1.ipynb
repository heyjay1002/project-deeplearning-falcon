```python
import time
import pandas as pd
from pathlib import Path

results = []
text = "KoreanAir 1153 Runway 25 wind 340 at 10 maximum 15 cleared to land."

# gTTS
from gtts import gTTS
start = time.time()
tts = gTTS(text=text, lang='en')
tts.save("gtts_output.mp3")
results.append({"TTS Engine": "gTTS", "Time (sec)": round(time.time() - start, 3)})

# Google Cloud TTS
from google.cloud import texttospeech
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "google-cloud-addinedu-dl-proj.json"
client = texttospeech.TextToSpeechClient()
synthesis_input = texttospeech.SynthesisInput(text=text)
voice = texttospeech.VoiceSelectionParams(language_code="en-US", ssml_gender=texttospeech.SsmlVoiceGender.NEUTRAL)
audio_config = texttospeech.AudioConfig(audio_encoding=texttospeech.AudioEncoding.MP3)

start = time.time()
response = client.synthesize_speech(input=synthesis_input, voice=voice, audio_config=audio_config)
Path("google_output.mp3").write_bytes(response.audio_content)
results.append({"TTS Engine": "Google Cloud TTS", "Time (sec)": round(time.time() - start, 3)})

# Coqui TTS
# from TTS.api import TTS
# tts_model = TTS(model_name="tts_models/en/ljspeech/vits")
# start = time.time()
# tts_model.tts_to_file(text=text, file_path="coqui_output.wav")
# results.append({"TTS Engine": "Coqui TTS", "Time (sec)": round(time.time() - start, 3)})

# 출력
import pandas as pd
print(pd.DataFrame(results).to_markdown(index=False))
```
